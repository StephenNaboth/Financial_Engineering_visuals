# Rolling Statistics

```python
data = pd.DataFrame(data["MSFT.O"])
```

```python
data.dropna().head(10)
```

```python
data.info()
```

```python
window = 10
```

```python
data["MSFT.O"].rolling(window=window).min().dropna().shape
```

```python
data['min'] = data["MSFT.O"].rolling(window=window).min()  
```

```python
data['min'] = data["MSFT.O"].rolling(window=window).min()  
```

```python
data['mean'] = data["MSFT.O"].rolling(window=window).mean() 
```

```python
data['std'] = data["MSFT.O"].rolling(window=window).std() 
```

```python
data['median'] = data["MSFT.O"].rolling(window=window).median() 
```

```python
data['max'] = data["MSFT.O"].rolling(window=window).max() 
```

```python
data['ewma'] = data["MSFT.O"].ewm(halflife=0.5, min_periods=window).mean() 
```

```python
data.dropna().head()
```

```python
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(
    figsize=(10, 6), style=['g--', 'r--', 'g--'], lw=0.8)  
data["MSFT.O"].iloc[-200:].plot(ax=ax, lw=2.0); 
```
